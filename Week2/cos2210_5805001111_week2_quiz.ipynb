**Quiz 2:**

* print all string methods using your name as their argument
* use comment or text to explain the outputs

```
my_name = "put pHapOn"
```

```
print(my_name.capitalize())

# input : string
# output : string ที่อักษรตัวแรกเป็น Uppercase
```

```
print(my_name.casefold())

# input : string
# output : คล้าย .ower()
```

```
print(my_name.center(100))

# input : จำนวน
# output : จัดรูป string ให้อยู่กึ่งกลางของจำนวนที่กำหนด
```

```
print(my_name.count('Put'))

# input : คำที่ต้องการนับ
# output : จำนวนคำ ที่อยู่ภายใน string
```

```
print(my_name.encode())

# input : string
# output : แปลง string ให้อยู่ในรูป UTF-8
```

```
print(my_name.endswith('pon'))

# input : string
# output : ตรวจสอบคำท้ายสตริง ว่าตรงตามที่ระบุหรือไม่
```

```
print(my_name.find('pHapOn'))

# input : string
# output : ตรวจสอบ index ของคำ ถ้า -1 จะแสดงว่าไม่พบ
```

```
print("* {} *".format(my_name))

# input : string
# output : จัดรูปแบบสตริงตามที่ปรับแต่งภายใน {}
```

```
p = {'x': my_name, 'y': '#'}
print("{y} {x} {y}".format_map(p))

# input : List, dict
# output : จัดรูปแบบสตริงตามที่ปรับแต่งภายใน {} โดยสามารถระบุ key ได้
```

```
print(my_name.index('pHapOn'))

# input : string
# output : ค้นหา Index แรก ของ subsrting
```

```
print(my_name.isalnum())

# input : string
# output : ตรวจสอบทั้ง string ว่ามีตัวเลขหรือไม่ (ถ้าพบ space จะให้ค่า False ทั้งที่)
```

```
print(my_name.isalpha())

# input : string
# output : ตรวจสอบทั้ง string ว่ามีตัวอักษรหรือไม่ (ถ้าพบ space จะให้ค่า False ทั้งที่)
```

```
print(my_name.isdecimal())

# input : string
# output : ตรวจสอบทั้ง string ว่ามีตัวเลขฐาน 10 หรือไม่ (ถ้าพบ space จะให้ค่า False ทั้งที่)
```

```
print(my_name.isdigit())

# input : string
# output : ตรวจสอบทั้ง string ว่ามีตัวเลขฐาน 2 หรือไม่ (ถ้าพบ space จะให้ค่า False ทั้งที่)
```

```
print(my_name.isidentifier())

# input : string
# output : ตรวจสอบ string ว่าถูกต้องตามกฎ identifier หรือไม่? (ไม่ขึ้นต้นด้วยตัวเลขหรือเว้นวรรค)
```

```
print(my_name.islower())

# input : string
# output : ตรวจสอบ string ว่าเป็นอักษรพิมพ์เล็กทั้งหมดหรือไม่?
```

```
print(my_name.isnumeric())

# input : string
# output : ตรวจสอบ string ว่าเป็นอักษรตัวเลขหรือไม่
```

```
print(my_name.isprintable())

# input : string
# output : ตรวจสอบ string ว่าสามารถพิมพ์ออกมาได้หรือไม่ (เช่นอักษร \r, \0)
```

```
print(my_name.isspace())

# input : string
# output : ตรวจสอบ string ว่าเป็นเว้นวรรคทั้งหมดหรือไม่? (อาจมี \t \n ได้)
```

```
print(my_name.istitle())

# input : string
# output : ตรวจสอบ string ว่ามีอักษรเพิมพ์ใหญ่เริ่มต้นหรือไม่?
```

```
print(my_name.isupper())

# input : string
# output : ตรวจสอบ string ว่าอักษรพิมพ์ใหญ่ทั้งหมดหรือไม่?
```
