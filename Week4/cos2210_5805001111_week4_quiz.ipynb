# ***Quiz 3:***
Depending on where an individual is from we need to tax them appropriately.
The states of CA, MN, and NY have taxes of 7.5%, 9.5%, and 8.9% respectively. Use
this information to take the amount of a purchase and the corresponding state to
assure that they are taxed by the right amount.

```
# Either
state_tax =  {
    'CA' : 0.075,
    'MN' : 0.095,
    'NY' : 0.089
}
```

```
# amount of purchase
purchase_amount = 4000
state = 'MN'

tax_amount = state_tax[state]
total_cost = purchase_amount*(1+tax_amount)

result = "Since you're from {}, your total cost is {}.".format(state, total_cost)
print(result)
```

# ***Quiz 8-1:***
Use the dictionary and list to count the total number of fruits

```
result = 0
basket_items = {'apples': 4, 'oranges': 19, 'kites': 3, 'sandwiches': 8}
fruits = ['apples', 'oranges', 'pears', 'peaches', 'grapes', 'bananas']

for i in fruits:
  if i in basket_items:
    print(basket_items[i])
    result += basket_items[i]

print("sum : " + str(result))
```

# ***Quiz 10:***
Write a loop with a `break` statement to create a string, `news_ticker`, that
is exactly 140 characters long.

```
headlines = ["Local Bear Eaten by Man",
"Legislature Announces New Laws",
"Peasant Discovers Violence Inherent in System",
"Cat Rescues Fireman Stuck in Tree",
"Brave Knight Runs Away",
"Papperbok Review: Totally Triffic"]
```

```
long_str = 0
news_ticker = ""

# write your loop here
for i in headlines:
  long_str += len(i)
  if long_str < 140:
    news_ticker += i
  else:
    break

print(news_ticker)
```

# ***Quiz 11:***
Use `zip` to write a `for` loop that creates a string specifying the label
and coordinates of each point and appends it to the list `points`. Each string
should be formatted as `label: x, y, z.` For example, the string for the
first coordinate should be `F: 23, 677, 4.`

```
x_coord = [23, 53, 2, -12, 95, 103, 14, -5]
y_coord = [677, 233, 405, 433, 905, 376, 432, 445]
z_coord = [4, 16, -6, -42, 3, -6, 23, -1]
labels = ["F", "J", "A", "Q", "Y", "B", "W", "X"]
points = []

for l, x, y, z in zip(labels,x_coord,y_coord,z_coord):
  points.append("{}: {}, {}, {}".format(l,x,y,z))

for point in points:
  print(point)
```
