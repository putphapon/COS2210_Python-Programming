## Variable

**Addition**

$c = a + b$

```
a = 1.0
b = 1.5
c = a+b
print(c)
```

**Subtraction**

$c = a − b$

```
a = 15
b = 39
c = a-b
print(c)
```

**Multiplication**

$c = a \times b$

$c = a  \cdot b$

```
a = 15
b = 39
c = a*b
print(c)
```

Division

$c =\frac{a}{b}$

```
a = 150
b = 10.0
c = a/b
print(c)
```

**Exponent**

$c = a^2 + b^3$

```
a = 1.5
b = 3
c = a**2 + b**3
print(c)
```

**Modulus**

$c = a \% b$

```
a = 15
b = 2
c = a%b
print(c)
```

## Variable type

```
# integer
x = 5
print(x)
print(type(x))
```

```
# float
y = 5.2
print(y)
print(type(y))
```

```
# cast int to float
x = 5
print(type(x))

x = float(x)
print(type(x))
```

```
# cast float to int
y = 5.2
print(type(y))

y = int(y)
print(type(y))
```

```
# int or float ?
print(4/2) # float
print(4//2) # int
print(4/2.0) # float
print(4.0/2) # float
```

```
# Boolean (True/False)
print(10-2 == 8) # True
```

```
# not equal
print(555 != 55)
```

```
# Case - sensitive
print('A' == 'a')
print('Hello' != "hello")
```

```
# variable assignment
my_bool = 2*2 == 4
my_bool
```

```
# True = 1, False = 0
x = True
print(x)
print(type(x))
```

```
# cast bool to int
x = int(x)
print(x)
print(type(x))
```

```
# cast int to bool
x = 0
print(x)
print(type(x))

x = bool(x)
print(x)
print(type(x))
```
