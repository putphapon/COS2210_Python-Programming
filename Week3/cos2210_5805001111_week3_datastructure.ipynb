# List

## Indexing

```
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October', 'November', 'December']

# From the front element
print(months[0])
print(months[1])
print(months[7])

# From the last element
print(months[-1])
print(months[-2])
```

## Slicing

```
# The third quater
q3 = months[6:9]
print(q3)
```

```
# From beginning element
first_half = months[:6]
print(first_half)
```

```
# Until the last element
second_half = months[6:]
print(second_half)
```

## Length of lists

```
greeting = "hello there"
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October', 'November', 'December']
```

```
print(len(greeting), len(months))
```

## Membership

```
# string
greeting = "hello there"
print('her' in greeting, 'her' not in greeting)
```

```
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October', 'November', 'December']
print('Sunday' in months, 'Sunday' not in months)
```

## Mutability and Order

```
# mutable
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October', 'November', 'December']
months[3] = 'Friday'
print(months)
```

```
# immutable
sentence = "I wish to register a complaint."
print(len(sentence))
sentence[30] = '!'
```

```
name = "Jim"
student = name
name = "Tim"

print(name)
print(student)
```

```
scores = ["B", "C", "A", "D", "B", "A"]

grades = scores
print("scores: "+ str(scores))
print("grades: "+ str(grades))
```

```
scores = ["B", "C", "A", "D", "B", "A"]

grades = scores
print("scores: "+ str(scores))
print("grades: "+ str(grades))

scores[3] = "B"
print("scores: "+ str(scores))
print("grades: "+ str(grades))
```

# Set

```
countries = ['Angola', 'Maldives', 'India', 'United States', 'India', 'Denmark', 'Sweden']
print(len(countries))
print(countries[:5])
```

```
country_set = set(countries)
print(len(country_set))
```

```
print(country_set[:3])
```

```
print('India' in countries)
print('India' in country_set)
```

```
country_set.add('Italy')
print(country_set)
```

# Dictionary

```
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}
print(elements['carbon'])
```

```
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}
elements['lithium'] = 3
print(elements)
```

```
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}
print(elements.get('dilithium'))
```

```
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}
print(elements['dilithium'])
```

```
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}
x = elements.get('dilithium')
is_null = x is None
print(is_null)
```

```
elements = {'hydrogen': 1, 'helium': 2, 'carbon': 6}
x = elements.get('dilithium')
is_null = x is not None
print(is_null)
```

# Compound Data Structures

```
elements = {"hydrogen": 
                            {"number": 1,
                              "weight": 1.00794,
                              "symbol": "H"},
                  "helium": 
                            {"number": 2,
                            "weight": 4.002602,
                            "symbol": "He"}}
```

```
print(elements['helium'])
print(elements.get('unobtainium', 'There\'s no such element!'))
```
