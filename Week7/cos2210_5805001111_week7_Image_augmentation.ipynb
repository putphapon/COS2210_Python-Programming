# Image augmentation

## Transformation

```python
# Required libraries
import numpy as np
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random

from skimage import exposure
from skimage.util import random_noise
from skimage.transform import rotate, AffineTransform, warp
from skimage.filters import gaussian
from cv2 import resize
```

## Upload an input image

```python
img = mpimg.imread('/content/domestic-cat-cat-adidas-relaxed.jpg')
```

## Resize (Scaling)

```python
img_rescale = resize(img,(400,400))
print("size of original image {}".format(img.shape))
print("size of scaled image {}".format(img_rescale.shape))

_ , (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img)
ax1.set_title("Original")
ax2.imshow(img_rescale)
ax2.set_title("Scale down")
plt.show()
```

## Flip - horizontal or vertical

```python
horiz = np.fliplr(img)
vert = np.flipud(img)

_, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(img)
ax1.set_title('original')
ax2.imshow(horiz)
ax2.set_title("Horizontal flip")
ax3.imshow(vert)
ax3.set_title("Verical flip")
plt.show()
```

## Rotation

```python
from skimage import transform

trans = rotate(img_rescale, random.uniform(-20,20))
trans_90 = rotate(img_rescale, random.uniform(80,90))

_, (ax1, ax2,  ax3) = plt.subplots(1, 3)
ax1.imshow(img)
ax1.set_title("Original")
ax2.imshow(trans)
ax2.set_title("Rotation")
ax3.imshow(trans_90)
ax3.set_title("Rotation 90*")
plt.show()
```

## shifiting images

```python
transform = AffineTransform(translation=(750,250))
wrapShift = warp(img,transform,mode='wrap')

_, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img)
ax1.set_title("Original")
ax2.imshow(wrapShift)
ax2.set_title("Wrap shift")
plt.show()
```

## Adding noise

```python
# random noise1
img_nos = random_noise(img_rescale, mode = 's&p', clip = True)

_, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img_rescale)
ax1.set_title("Original")
ax2.imshow(img_nos)
ax2.set_title("noise injection")
plt.show()
```

```python
#ramdom noise 2
#standard deviation for noise to be added in the image
sigma = 0.255

#add random noise to the image
noisyRandom = random_noise(img_rescale, var=sigma**2)

_, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img_rescale)
ax1.set_title("Original")
ax2.imshow(noisyRandom)
ax2.set_title("noise injection")
plt.show()
```

## Blurring image

```python
blurred = gaussian(img_rescale, sigma=10, multichannel=True)

_, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img_rescale)
ax1.set_title("Original")
ax2.imshow(blurred)
ax2.set_title("Blurred image")
plt.show()
plt.show()
```
